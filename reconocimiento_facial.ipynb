```python
import cv2
import os
import shutil
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

params = {
    "imagePath": "d/MyDrive/faceDetection/image.jpg",
    "cascadePath": "d/MyDrive/faceDetection/haarcascade_frontalface_default.xml"
}
```

# STATIC PHOTO

```python
# get cv cascade 
shutil.copy(params["cascadePath"], "haarcascade_frontalface_default.xml")
face_—Åascade = cv2.CascadeClassifier(
    os.path.join(
        cv2.data.haarcascades,
        "haarcascade_frontalface_default.xml"
    )
)
```

```python
# get image
img = cv2.imread(params["imagePath"])
plt.imshow(img);
```

```python
# switch image to grayscale
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img);
```

```python
faces = face_—Åascade.detectMultiScale(
    img,
    scaleFactor=1.5, # compensates objects' diff distance
    minNeighbors=5, # objects detected near moving window
    minSize=(40, 40), # moving window size
    flags = cv2.CASCADE_SCALE_IMAGE
)
print(f"no of the objects detected: {len(faces)}")
print(faces)
```

```python
# get (x, y, width, height) of the rectangle
for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0, 255, 0), 2)
cv2_imshow(img)
```
